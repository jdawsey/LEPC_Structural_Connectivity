```python
merged_data = pd.read_csv("E:/!!Research/!!!Data/graph_analysis/igraph_metrics_data/all_leks_metrics/max_dispersal_metrics_joined.csv")

# Data preparation
def prepare_centrality_data(merged_data):
    # Filter only surveyed leks
    surveyed_leks = merged_data[merged_data['activity'].isin(['TRUE', 'FALSE'])]
    
    # Group by year and activity, calculate mean centrality
    activity_centrality = surveyed_leks.groupby(['year', 'activity'])['centrality_score'].mean().unstack()
    
    return activity_centrality

# Visualization
def plot_lek_centrality(activity_centrality):
    plt.figure(figsize=(10, 6))
    
    # Create bar plot
    activity_centrality.plot(kind='bar', ax=plt.gca())
    
    plt.title('Mean Centrality Scores by Lek Activity and Year')
    plt.xlabel('Year')
    plt.ylabel('Mean Centrality Score')
    plt.legend(title='Lek Status')
    plt.tight_layout()
    plt.show()

# Boxplot for distribution
def plot_lek_centrality_distribution(merged_data):
    plt.figure(figsize=(10, 6))
    
    # Filter only surveyed leks
    surveyed_leks = merged_data[merged_data['activity'].isin(['TRUE', 'FALSE'])]
    
    # Create boxplot
    sns.boxplot(x='year', y='centrality_score', hue='activity', data=surveyed_leks)
    
    plt.title('Centrality Score Distribution by Lek Activity and Year')
    plt.xlabel('Year')
    plt.ylabel('Centrality Score')
    plt.tight_layout()
    plt.show()

# Example usage
plot_lek_centrality(prepare_centrality_data(merged_data))
plot_lek_centrality_distribution(merged_data)
```

```python
def plot_lek_centrality_line(merged_data):
    # Filter only surveyed leks
    surveyed_leks = merged_data[merged_data['activity'].isin(['TRUE', 'FALSE', 'not surveyed'])]
    
    # Group by year and activity, calculate mean and standard error of centrality scores
    centrality_summary = surveyed_leks.groupby(['year', 'activity'])['centrality_score'].agg(['mean', 'sem', 'count']).reset_index()
    
    # Separate active and inactive data
    active_data = centrality_summary[centrality_summary['activity'] == 'TRUE']
    inactive_data = centrality_summary[centrality_summary['activity'] == 'FALSE']
    not_surveyed_data = centrality_summary[centrality_summary['activity'] == 'not surveyed']
    
    # Create the plot
    plt.figure(figsize=(10, 6))
    
    # Plot active leks with filled confidence interval
    plt.plot(active_data['year'], active_data['mean'], '-o', color='yellow', label='Active Leks')
    plt.fill_between(active_data['year'], 
                     active_data['mean'] - active_data['sem'], 
                     active_data['mean'] + active_data['sem'], 
                     color='green', alpha=0.2)
    
    # Plot inactive leks with filled confidence interval
    plt.plot(inactive_data['year'], inactive_data['mean'], '-o', color='red', label='Inactive Leks')
    plt.fill_between(inactive_data['year'], 
                     inactive_data['mean'] - inactive_data['sem'], 
                     inactive_data['mean'] + inactive_data['sem'], 
                     color='red', alpha=0.2)
    
    # Plot not surveyed leks with filled confidence interval
    plt.plot(not_surveyed_data['year'], not_surveyed_data['mean'], '-o', color='blue', label='Not surveyed Leks')
    plt.fill_between(not_surveyed_data['year'], 
                     not_surveyed_data['mean'] - not_surveyed_data['sem'], 
                     not_surveyed_data['mean'] + not_surveyed_data['sem'], 
                     color='red', alpha=0.2)
    
    plt.title('Mean Centrality Scores by Lek Activity and Year')
    plt.xlabel('Year')
    plt.ylabel('Mean Centrality Score')
    plt.legend(title='Lek Status')
    plt.grid(True, linestyle='--', alpha=0.7)
    plt.tight_layout()
    plt.show()

# Usage
plot_lek_centrality_line(merged_data)
```

```python
def plot_lek_centrality_percentile(merged_data):
    # Filter only surveyed leks
    surveyed_leks = merged_data[merged_data['activity'].isin(['TRUE', 'FALSE'])]
    
    # Group by year and activity, calculate 90th percentile of centrality scores
    centrality_summary = surveyed_leks.groupby(['year', 'activity'])['centrality_score'].quantile(0.9).reset_index()
    centrality_summary.columns = ['year', 'activity', 'percentile_90']
    
    # Separate active and inactive data
    active_data = centrality_summary[centrality_summary['activity'] == 'TRUE']
    inactive_data = centrality_summary[centrality_summary['activity'] == 'FALSE']
    
    # Create the plot
    plt.figure(figsize=(10, 6))
    
    # Plot active leks
    plt.plot(active_data['year'], active_data['percentile_90'], '-o', color='yellow', label='Active Leks')
    plt.fill_between(active_data['year'], 
                     active_data['percentile_90'], 
                     np.min(active_data['percentile_90']),
                     color='green', alpha=0.2)
    
    # Plot inactive leks
    plt.plot(inactive_data['year'], inactive_data['percentile_90'], '-o', color='red', label='Inactive Leks')
    plt.fill_between(inactive_data['year'], 
                     inactive_data['percentile_90'], 
                     np.min(inactive_data['percentile_90']),
                     color='red', alpha=0.2)
    
    plt.title('90th Percentile of Centrality Scores by Lek Activity and Year')
    plt.xlabel('Year')
    plt.ylabel('90th Percentile Centrality Score')
    plt.legend(title='Lek Status')
    plt.grid(True, linestyle='--', alpha=0.7)
    plt.tight_layout()
    plt.show()

# Usage
plot_lek_centrality_percentile(merged_data)
```
