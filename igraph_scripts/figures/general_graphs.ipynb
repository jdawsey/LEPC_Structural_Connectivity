Used to generate graphs related to general descriptions of leks

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = "E:/!!Research/!!!Data/graph_analysis/lek_data/lek_data_survey_tally.csv"
df = pd.read_csv(data)
df.head()
```

```python
# Set the style
sns.set_style("white")

# Choose the column to plot
column = 'cumulative_leks_surveyed'

# Prepare data for stackplot
years = df['year']
cumulative_values = df[column] #.cumsum()  # Compute cumulative total for the column

# Create the plot
fig, ax = plt.subplots(figsize=(10, 6))
ax.fill_between(years, cumulative_values, label=column, alpha=0.8)

# Add labels and legend
#ax.set_title(f'Stacked Area Plot for cumulative leks surveyed', fontsize=16)
ax.set_xlabel('Year', fontsize=14)
ax.set_ylabel('Cumulative leks surveyed', fontsize=14)
# ax.legend(loc='upper left')

# Save the plot
plot_folder = "E:/!!Research/!!Documents/Documents_2024/manuscriptDraft24_connectivity/figures/graphs"
plt.savefig(f'{plot_folder}/cumulative_leks.png', dpi=300, bbox_inches = 'tight')

plt.tight_layout()
plt.show()
```

```python
# Prepare data
years = df['year']
cumulative_values = df['cumulative_leks_surveyed']  # Cumulative totals
leks_surveyed = df['leks_surveyed_this_year']  # Number of leks surveyed

# Create the plot
fig, ax1 = plt.subplots(figsize=(10, 6))

# Plot the stacked area for cumulative totals
ax1.fill_between(years, cumulative_values, label='Cumulative leks surveyed', alpha=0.8, color='skyblue')
ax1.set_xlabel('Year', fontsize=14)
ax1.set_ylabel('Cumulative Total', fontsize=14)
ax1.tick_params(axis='y')

# Overlay the line plot for leks surveyed (secondary axis)
ax2 = ax1.twinx()  # Create a secondary y-axis
ax2.plot(years, leks_surveyed, label='Leks surveyed per year', color='darkorange', marker='o', linewidth=2)
ax2.set_ylabel('Leks Surveyed', fontsize=14)
ax2.tick_params(axis='y')

# Add legends and title
#fig.suptitle('Cumulative Totals and Leks Surveyed Over Years', fontsize=16)
fig.legend(loc='upper left', bbox_to_anchor=(0.1, 0.9), bbox_transform=ax1.transAxes)

# Save and display the plot
plt.tight_layout()

# Save the plot
plot_folder = "E:/!!Research/!!Documents/Documents_2024/manuscriptDraft24_connectivity/figures/graphs"
plt.savefig(f'{plot_folder}/cumulative_leks_vs_surveyed.png', dpi=300, bbox_inches = 'tight')
plt.show()
```
