```python
import pandas as pd
```

```python
df = pd.read_csv("E:/!!Research/!!!Data/graph_analysis/lek_data/raw_data/lek_data_binary_year_activity_for_py.csv")
df.head()
```

```python
# Group by 'site' and 'year', and aggregate by taking the max for 'activity'
df_result = df.groupby(['lek_id', 'year'], as_index=False).agg({'activity': 'max',
                                                                'x_easting' : 'first',
                                                                'y_northing' : 'first'})

df_result.head()
```

```python
# Load the Excel file into a pandas DataFrame
file_path = "E:/!!Research/!!!Data/graph_analysis/lek_data/cleaned_data/lek_data_binary_cleaned_redo.csv"  # Replace with the path to your Excel file
df = pd.read_csv(file_path)

# Find duplicates based on coordinates (northings and eastings) but keep all relevant rows
duplicates_by_coords = df[df.duplicated(subset=['y_northing', 'x_easting'], keep=False)]

# Group these duplicates by coordinates to check their differences
grouped = duplicates_by_coords.groupby(['y_northing', 'x_easting'])

# Create a summary DataFrame for review
summary = []
for coords, group in grouped:
    summary.append({
        'y_northing': coords[0],
        'x_easting': coords[1],
        'lek_id': ', '.join(map(str, group['lek_id'].unique())),
        'year': ', '.join(map(str, sorted(group['year'].unique()))),
        'activity': ', '.join(map(str, group['activity'].unique()))
    })

# Convert summary to a DataFrame
summary_df = pd.DataFrame(summary)

# Export the results to CSV
duplicates_by_coords.to_csv("E:/!!Research/!!!Data/graph_analysis/lek_data/cleaned_data/lek_data_binary_duplicate_rows.csv", index=False)  # All duplicate rows
summary_df.to_csv("E:/!!Research/!!!Data/graph_analysis/lek_data/cleaned_data/lek_data_binary_coordinate_duplicates_summary.csv", index=False)  # Summary file

print("Results saved to 'duplicate_rows.csv' and 'coordinate_duplicates_summary.csv'")
```
