Have to change spreadsheet to "lek" column becomes "lek_id"

Max dispersal

```python
# Read the CSV files
# CSV with ids, northings, eastings, centrality, hub scores
lek_scores = pd.read_csv("E:/!!Research/!!!Data/graph_analysis/igraph_metrics_data/all_leks_metrics/max_dispersal_metrics.csv")
# Spreadsheet with yearly lek info
yearly_lek_data = pd.read_csv("E:/!!Research/!!!Data/graph_analysis/lek_data/lek_data_binary_rolling_activity.csv")

# Join the dataframes based on a common identifier
# Assuming the common column is 'id' in both dataframes
merged_data = yearly_lek_data.merge(lek_scores[['lek_id', 'y_northing', 'x_easting', 'centrality_score', 'hub_score']], 
                                     on='lek_id', 
                                     how='left')

merged_data.to_csv("E:/!!Research/!!!Data/graph_analysis/igraph_metrics_data/all_leks_metrics/max_dispersal_metrics_joined.csv")
merged_data.tail()
```
