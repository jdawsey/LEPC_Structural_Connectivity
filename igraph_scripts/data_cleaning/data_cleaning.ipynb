```python
import pandas as pd
```

```python
df = pd.read_csv("E:/!!Research/!!!Data/graph_analysis/lek_data/lek_data_binary_year_activity_for_py.csv")
df.head()
```

```python
# Group by 'site' and 'year', and aggregate by taking the max for 'activity'
df_result = df.groupby(['lek_id', 'year'], as_index=False).agg({'activity': 'max',
                                                                'x_easting' : 'first',
                                                                'y_northing' : 'first'})

df_result.head()
```

```python
# Count unique years for each site
df_result['year_count'] = df_result.groupby('lek_id')['year'].transform('nunique')

# Filter to keep only sites with more than two years
df_filtered = df_result[df_result['year_count'] > 2].drop(columns=['year_count'])

df_filtered.head()
```

```python
df_filtered.to_csv("E:/!!Research/!!!Data/graph_analysis/lek_data/lek_data_binary_cleaned.csv")
```
