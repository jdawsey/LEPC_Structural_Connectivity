```python
import pandas as pd
```

```python
csv_path = "E:/!!Research/!!!Data/graph_analysis/igraph_metrics_data/lek_rolling_binary_node_metrics.csv"
df = pd.read_csv(csv_path)
df.head()
```

Finding hubs

Average dispersal hubs

```python
df_ad = df[df['graph_type'] == 'average_dispersal']


# Define threshold for identifying hubs
threshold = 0.9

# Step 1: Add a column indicating if the lek is a hub for each year
df_ad['is_hub'] = df_ad['hub_score'] >= threshold

# Step 2: Count how many times each lek has been a hub
hub_counts_ad = df_ad.groupby(['lek', 'x_easting', 'y_northing'])['is_hub'].sum().reset_index()

# Rename the columns for clarity
hub_counts_ad.columns = ['lek', 'x_easting', 'y_northing', 'hub_count']

# Step 3: Save to CSV
hub_counts_ad.to_csv('E:/!!Research/!!!Data/graph_analysis/igraph_metrics_data/lek_hub_counts_average_dispersal.csv', index=False)

hub_counts_ad.tail()
```

Testing normalization by count

```python
df_ad_n = df[df['graph_type'] == 'average_dispersal']


# Define threshold for identifying hubs
threshold = 0.9

# Step 1: Add a column indicating if the lek is a hub for each year
df_ad_n['is_hub'] = df_ad_n['hub_score'] >= threshold

# Step 2: Count how many times each lek has been a hub and the number of active years (rows per lek)
hub_counts_ad_n = df_ad_n.groupby(['lek', 'x_easting', 'y_northing']).agg(
    hub_count=('is_hub', 'sum'),
    active_years=('hub_score', 'count')  # Counts each row, indicating active years
).reset_index()


# Step 3: Save to CSV
hub_counts_ad_n.to_csv('E:/!!Research/!!!Data/graph_analysis/igraph_metrics_data/lek_hub_counts_average_dispersal_normal.csv', index=False)

hub_counts_ad_n.tail()
```

### Starting with betweenness centrality

Average dispersal

```python
df_bc_ad = df[df['graph_type'] == 'average_dispersal']


# Define threshold for identifying hubs
threshold = 0.9

# Calculate the 90th quantile of betweenness_centrality across all data
quantile_90 = df_bc_ad['centrality_score'].quantile(0.9)

# Step 1: Add a column indicating if the lek is a hub for each year and if it exceeds the 90th quantile
df_bc_ad['is_stepping_stone'] = df_bc_ad['centrality_score'] >= threshold
df_bc_ad['above_90th_quantile'] = df_bc_ad['centrality_score'] > quantile_90

# Step 2: Count how many times each lek has been a hub, the number of active years, and exceedances of the 90th quantile
hub_counts_bc_ad = df_bc_ad.groupby(['lek', 'x_easting', 'y_northing']).agg(
    hub_count=('is_stepping_stone', 'sum'),
    active_years=('centrality_score', 'count'),  # Counts active years
    above_90th_count=('above_90th_quantile', 'sum')  # Counts years above the 90th quantile
).reset_index()


# Step 3: Save to CSV
hub_counts_bc_ad.to_csv('E:/!!Research/!!!Data/graph_analysis/igraph_metrics_data/lek_step_stone_counts_average_dispersal.csv', index=False)

hub_counts_bc_ad.tail()
```
